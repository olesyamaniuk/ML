```python
import numpy as np
import time

a1 = np.random.random((200,200));
a2 = np.random.random((200,200));
```

```python
starttime = time.time()
result1= np.dot(a1,a2) 
endtime = time.time()

print("Time with numpy package:" + str(endtime - starttime))
```

```python
def MMultiply(x,y):
    result = np.zeros((200,200))
    for i in range(len(x)):
        for j in range(len(y[0])):
            for k in range(len(y)):
                result[i][j] += x[i][k] * y[k][j]
    return result
```

```python
starttime2 = time.time()
result2= MMultiply(a1, a2)
endtime2 = time.time()

print("Time with function:" + str(endtime2 - starttime2))
```
